Goal:

To analyze U.S Airbnb rentals data in order to increase understanding of patterns in the lodging and rentals industry and how they relate to demographics, economic, and geographic trends at large.

Business Problem:

Some resident and politicians of high cost of living (HCOL) areas have had concerns that the presence of short term rental platforms, such as Airbnb, have made their city less affordable for middle class residents. Arguments posed state that the advantage of short term rentals have benefited wealthy individuals with the means of becoming Airbnb investors. In return real estate markets have been comoditised to a point where most living options (rent & buy) have increasingly become unafordable.

This study will seek to study weather this phenomenon is true by analyzing Airbnb and US real estate data between 2016 and 2018.  These dates were selected to investigate the effect of Airbnb on real estate pricing by exploiting a natural experiment derived
from laws restricting short term rentals in some cities. Cities such as New York, Chicago and Denver enacted these laws in 2016 while other cities only began regulating short-term rentals later: Washington DC in 2018, and Boston in 2019.
